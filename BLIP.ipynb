```python
from datasets import load_dataset
from sklearn.model_selection import train_test_split

# Load the ROCO-radiology dataset
dataset = load_dataset("mdwiratathya/ROCO-radiology")
```

```python
# Use only 20% of the entire dataset
train_val_data = dataset['train']
train_val_data = train_val_data.shuffle(seed=42)

# Get 20% of the total data for training and testing
small_dataset = train_val_data.select(range(int(0.5 * len(train_val_data))))
print(small_dataset)
```

```python
# Convert small dataset to a list of dictionaries for easier manipulation
small_data_list = [{"image": item["image"], "caption": item["caption"]} for item in small_dataset]

# Split the small dataset into train and test sets (50% train, 50% test)
train_data, test_data = train_test_split(small_data_list, test_size=0.5, random_state=42)

# Show a preview of the data
print(f"Training set size: {len(train_data)}")
print(f"Testing set size: {len(test_data)}")
```

```python
# Extract images and captions from the train and test data
train_images = [item['image'] for item in train_data]
train_captions = [item['caption'] for item in train_data]

test_images = [item['image'] for item in test_data]
test_captions = [item['caption'] for item in test_data]

# Create training and testing datasets in a structured format
train_pairs = list(zip(train_images, train_captions))
test_pairs = list(zip(test_images, test_captions))

# Verify the image-caption pair extraction
print(f"Sample training image: {train_images[0]}")
print(f"Sample training caption: {train_captions[0]}")
```

```python
from transformers import BlipProcessor
from torch.utils.data import Dataset, DataLoader

# Load the processor for BLIP
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")

# Custom Dataset Class
class ImageCaptionDataset(Dataset):
    def __init__(self, image_paths, captions, processor, max_length=32):
        self.image_paths = image_paths
        self.captions = captions
        self.processor = processor
        self.max_length = max_length

    def __len__(self):
        return len(self.captions)

    def __getitem__(self, idx):
        image = self.image_paths[idx]
        caption = self.captions[idx]

        # Process the image
        pixel_values = self.processor(images=image, return_tensors="pt").pixel_values.squeeze(0)

        # Process the caption
        inputs = self.processor(text=caption, return_tensors="pt", padding="max_length", truncation=True, max_length=self.max_length)

        return {
            "input_ids": inputs.input_ids.squeeze(0),  # shape [seq_length]
            "attention_mask": inputs.attention_mask.squeeze(0),  # shape [seq_length]
            "pixel_values": pixel_values,  # shape [3, height, width]
        }

# Assuming you have train_images, train_captions, test_images, and test_captions ready

# Create Dataset and DataLoader for training
train_dataset = ImageCaptionDataset(train_images, train_captions, processor)
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)

# Create Dataset and DataLoader for testing
test_dataset = ImageCaptionDataset(test_images, test_captions, processor)
test_loader = DataLoader(test_dataset, batch_size=4, shuffle=False)

# Now you can print the number of training and testing batches
print(f"Number of training batches: {len(train_loader)}")
print(f"Number of testing batches: {len(test_loader)}")

# Check the DataLoader output for training data
for batch in train_loader:
    print("Input IDs shape:", batch["input_ids"].shape)
    print("Attention Mask shape:", batch["attention_mask"].shape)
    print("Pixel Values shape:", batch["pixel_values"].shape)
    break  # Just check the first batch

# Check the DataLoader output for testing data
for batch in test_loader:
    print("Input IDs shape:", batch["input_ids"].shape)
    print("Attention Mask shape:", batch["attention_mask"].shape)
    print("Pixel Values shape:", batch["pixel_values"].shape)
    break  # Just check the first batch
```

```python
from transformers import BlipForConditionalGeneration
import torch
from torch.optim import AdamW

# Load BLIP model for image captioning
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Move the model to the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Optimizer
optimizer = AdamW(model.parameters(), lr=5e-5)

# Training loop
epochs = 5  # Set the number of epochs
for epoch in range(epochs):
    model.train()
    total_loss = 0

    for batch in train_loader:
        # Move inputs to device
        input_ids = batch["input_ids"].to(device)
        pixel_values = batch["pixel_values"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        # Forward pass
        outputs = model(input_ids=input_ids, pixel_values=pixel_values, attention_mask=attention_mask, labels=input_ids)
        loss = outputs.loss
        total_loss += loss.item()

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    avg_loss = total_loss / len(train_loader)
    print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}")
```

```python
# Initialize total loss and metrics
total_loss = 0
predictions = []
references = []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch["input_ids"].to(device)
        pixel_values = batch["pixel_values"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        # Forward pass
        outputs = model(input_ids=input_ids, pixel_values=pixel_values, attention_mask=attention_mask, labels=input_ids)
        loss = outputs.loss
        total_loss += loss.item()

        # Generate predictions - set max_length for consistency
        generated_ids = model.generate(pixel_values=pixel_values, attention_mask=attention_mask, max_length=32, num_beams=1)


        # Decode the predictions and references
        predictions.extend(processor.batch_decode(generated_ids, skip_special_tokens=True))
        references.extend(processor.batch_decode(input_ids, skip_special_tokens=True))

# Calculate average loss
avg_loss = total_loss / len(test_loader)
print(f"Test Loss: {avg_loss:.4f}")

# Calculate BLEU score
bleu_score = metric.compute(predictions=predictions, references=[[ref] for ref in references])
print(f"BLEU Score: {bleu_score['bleu']:.4f}")
```
